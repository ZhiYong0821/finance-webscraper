# Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.2'
}

url = 'https://www.slickcharts.com/sp500'

page = requests.get(url=url, headers=headers)

soup = BeautifulSoup(page.text, 'html')

soup
```

```python
table = soup.find_all('table')[0]

table
```

```python
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

headers
```

```python
rows = []
for tr in table.find_all('tr')[1:]:
    cells = tr.find_all('td')
    row = [cell.text.strip() for cell in cells]
    rows.append(row)

rows
```

```python
df = pd.DataFrame(rows, columns=headers)

df
```

```python
df['% Chg'] = df['% Chg'].str.replace(r'[%()]', '', regex = True).astype(float)

df_sorted = df.sort_values(by='% Chg', ascending = False)

df_sorted
```
